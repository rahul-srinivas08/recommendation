```python
import numpy as np
import pandas as pd
```

```python
movie_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credit_df = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credit_df.head(3)
```

```python
movie_df.head(3)
```

```python
movie_df.shape
```

```python
credit_df.shape
```

```python
credit_rename = credit_df.rename(index = str , columns = {'movie_id':'id'})
movie_df_merge = movie_df.merge(credit_rename , on = 'id')
movie_df_merge.head()
```

```python
merge_cleaned_df = movie_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
merge_cleaned_df.head()
```

```python
merge_cleaned_df.info()
```

```python
#weight Average
V = merge_cleaned_df["vote_count"]
R = merge_cleaned_df["vote_average"]
c = merge_cleaned_df['vote_average'].mean()
print(c)
m = merge_cleaned_df["vote_count"].quantile(0.70)
```

```python
merge_cleaned_df['weighted_average']=((R*V)+ (c*m))/(V+m)
```

```python
merge_cleaned_df['weighted_average']
```

```python
merge_movie_sort = merge_cleaned_df.sort_values('weighted_average',ascending=False)
merge_movie_sort[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
weight_average=merge_movie_sort.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
```

```python
##Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movies_scaled_df = scaling.fit_transform(merge_cleaned_df[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movies_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
merge_cleaned_df[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
merge_cleaned_df.head()
```

```python
merge_cleaned_df['score'] = merge_cleaned_df['normalized_weight_average'] * 0.5 + merge_cleaned_df['normalized_popularity'] * 0.5
movies_scored_df = merge_cleaned_df.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```
